Importing libraries that are important for the model building and training

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report,accuracy_score
```

Data loading

```python
df=pd.read_csv("parkinsons.csv")
```

```python
df
```

Feature Extraction

```python
df.columns
```

Here we considered only important columns form the data set and stored in variable train

```python
train=df[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)','MDVP:Shimmer','NHR', 'HNR','RPDE', 'DFA','PPE','status']]
```

```python
train
```

```python
#Here their is no null values in the data set
train.isna().sum() 
```

```python
#the given data is unbiased
train["status"].value_counts()
```

```python
train.columns
```

```python
#Finding the input and output columns and storinng it in different  variables
#for input
x=train[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Shimmer', 'NHR', 'HNR', 'RPDE', 'DFA', 'PPE']]

#for output
y=train['status']
```

```python
print(x)
print(y)
```

```python
#As we observed that the given data is unbalanced 
#so to resolve it we go for over sampling method

from imblearn.over_sampling import RandomOverSampler
```

```python
#model calling for over sampling
ros=RandomOverSampler()
x_resample,y_resample=ros.fit_resample(x,y)
```

```python
len(x_resample)
```

```python
len(y_resample)
```

```python
#Now the data is resampled and balanced
y_resample.value_counts()
```

```python
#Spliting  the data into 2 parts for training and testing

x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=188)
```

```python
#model calling

model=GaussianNB()
```

```python
#Training  the model
model.fit(x_train,y_train)
```

```python
#Predin=cting the output or y_pred
y_pred=model.predict(x_test)
```

```python
#with  the help of the for loop finding the best  random state value for better accuracy

l=[]
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=i)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accu=accuracy_score(y_test,y_pred)
    l.append(accu)
    print(f'for random state {i} accuracy is {accu}')

print(f'best random state is {l.index(max(l))} with thhe accuracy of {max(l)}')
```

```python
#Calculating the accuracy

print(classification_report(y_test,y_pred))
```

Loading and storing the built model for parkinsons disease detection
